```python
import numpy as np
import bokeh
import cairosvg
import imageio
from bokeh.plotting import figure, output_notebook, show
```

```python
w, h = 800, 600
cairosvg.svg2png(url='field.svg', write_to="field.png", output_width=w, output_height=h)
cairosvg.svg2png(url='https://logotyp.us/file/nfl.svg', write_to="nfl.png")

img = imageio.imread('field.png')
img = img[::-1, :, :]
logo = imageio.imread('nfl.png')
logo = logo[::-1, :, :]

output_notebook()

p = figure(plot_width=w, plot_height=h, x_range=(0, w), y_range=(0, h))


p.image_rgba(image=[img], x=[0], y=[0], dw=[w], dh=[h])
logow, logoh = 56, 40
p.image_rgba(image=[logo], x=[w//2 - (logow//2)], y=[h//2 - (logoh//2)], dw=[logow], dh=[logoh])

show(p)
```
